```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = df.values
X = data[:,1:]
y = data[:,0]
```

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.2,random_state=42)
```

```python
cov_mat = np.cov(X_train.T)
eigen_vals, eigen_vecs = np.linalg.eig(cov_mat)
```

```python
print(eigen_vals)
```

```python
print(eigen_vecs)
```

```python
tot = sum(eigen_vals)
sorted_eigen_vals = sorted(eigen_vals, reverse=True)
var_exp = [i/tot for i in sorted_eigen_vals]
cum_var_exp = np.cumsum(var_exp)
```

```python
var_exp
```

```python
cum_var_exp
```

```python
plt.bar(range(1,14),var_exp,label = 'Individual Explained Variance Ratio')
plt.plot(range(1,14), cum_var_exp, label = 'Cummulative Explained Variance Ratio')
plt.legend()
plt.xlabel('Index')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

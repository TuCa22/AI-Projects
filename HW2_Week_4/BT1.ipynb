Nạp thư viện

```python
import numpy as np
import matplotlib.pyplot as plt
```

Tạo dữ liệu ngẫu nhiên có dạng tuyến tính: $y = 4 + 2x$

```python
np.random.seed(42)
x = 2 * np.random.rand(100,1)
y = 2 + 2 * x + np.random.randn(100,1)
plt.plot(x,y,'o')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Sử dụng phương pháp Gradient Descent

```python
x1 = np.c_[np.ones((100,1)),x]
alpha = 0.1 #Learning rate
n_interation = 1000
m = len(y) #Tổng số phần tử

w_gd = np.random.randn(2,1)    #Tạo giá trị đầu
w_gd_list = []

for i in range(n_interation):
    gradient = 2 / m * x1.transpose().dot(x1.dot(w_gd)-y)
    w_gd = w_gd - alpha*gradient
    w_gd_list.append(w_gd)
    y1 = w_gd[1] * x1 + w_gd[0]
    plt.plot(x1,y1)
    plt.plot(x,y,'o')
```

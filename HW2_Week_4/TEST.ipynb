# 2. Feature selections
- Choose features to train ML model
- Need to use "Feature Engineering" to identify features needed

```python
features = ['LotArea','YearBuilt','1stFlrSF','2ndFlrSF','FullBath','BedroomAbvGr','TotRmsAbvGrd']
```

# 3. Spliting datasets
- "data": dataset
- "X": data[features]
- "y": target variable "SalePrice"

```python
X = data[features]
y = data['SalePrice']
data.head()
```

```python
X.head()
```

```python
y.head()
```

X, y --> X_train, y_train, X_test, y_test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

# 4. Training machine learning model

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=1)
```

- Fit training data into model

```python
dt_model.fit(X_train,y_train)
```

- Predict

```python
y_pred = dt_model.predict(X_test.head())
```

```python
y_pred
```

```python
pd.DataFrame({'y': y_test.head(), 'y_pred': y_pred})
```

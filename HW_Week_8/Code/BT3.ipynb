```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[2, 2],[3, 3],[4, 4],[5, 5],[4, 6],[3, 7],[4, 8],[5, 9],[6, 10]])
B = np.array([[6, 2],[7, 3],[8, 4],[9, 5],[8, 6],[7, 7],[7, 8],[7, 9],[8, 10]])

```

```python
x1 = A[:,0]
y1 = A[:,1]
x2 = B[:,0]
y2 = B[:,1]
plt.scatter(x1, y1, c = 'red', label = 'Class -1')
plt.scatter(x2, y2, c = 'green', label = 'Class +1')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Phân bố dữ liệu trên đồ thị 2D')
plt.legend()
plt.show()
```

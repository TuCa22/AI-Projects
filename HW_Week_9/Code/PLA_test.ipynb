```python
import numpy as np
import pandas as pd

# X là ma trận dữ liệu đầu vào, y là ma trận nhãn (labels)
iris = pd.read_csv('D:/Programming/Python/Projects/HW_Week_9/Data/iris.csv')
X_iris = iris[['sepal.length','sepal.width']]
y_iris = iris['variety']

X = X_iris.values
y0 = y_iris.values
```

```python
# Khởi tạo trọng số w và bias b ban đầu
w = np.zeros(X.shape[1])
b = 0

# Tham số learning rate
learning_rate = 0.1

# Số lần lặp (epochs)
epochs = 10
```

```python
#Quy đổi label dạng chuỗi sang số nguyên
label_mapping = {'Setosa':0,'Versicolor':1,'Virginica':2}
# labels = {'Setosa','Versicolor','Virginica'}
y = [label_mapping[i] for i in y0]
```

```python
# Training Perceptron
for epoch in range(epochs):
    for i in range(X.shape[0]):
        xi = X[i]
        yi = y[i]
        z = np.dot(w, xi) + b
        prediction = 1 if z >= 0 else 0
        error = yi - prediction
        w += learning_rate * error * xi
        b += learning_rate * error
```

```python
# Kiểm tra mô hình đã học
for i in range(X.shape[0]):
    xi = X[i]
    z = np.dot(w, xi) + b
    prediction = 1 if z >= 0 else 0
    print(f"Input: {xi}, Prediction: {prediction}")
```
